```r
install.packages('TurtleGraphics')
library('TurtleGraphics')
turtle_init()

get_color <- function(color) {
    if (color == 'red') {
        r = 1
        g = 0
        b = 0
        return (r, g, b)
    } else if (color == 'blue') {
        r = 0
        g = 0
        b = 1
        return (r, g, b)
    } else if (color == 'white') {
        r = 1
        g = 1
        b = 1
        return (r, g, b)
    } else {
        r = 0
        g = 0
        b = 0
        return (r, g, b)
    }
}

# Turtle in R does not allow a color to be filled 
# I commented out the fillcolor command and replaced it with turtle_col 
# I already talked to Professor Fang about this 

draw_rectangle <- function(length, height, color) {
    turtle_down()
    #turtle_fillcolor(getcolor(color))
    turtle_col(col=color)
    for (i in 1:4) {
        if (i == 1 | i == 3) { 
            turtle_forward(dist=length)
            turtle_right(90)
        } else {
            turtle_forward(dist=height)
            turtle_right(90)
        }
    }
    #turtle_endfill()
    turtle_up()
}

draw_rectangle(10, 10, 'red')
# Turtle does not execute well in notebook but it performs the function 
```
